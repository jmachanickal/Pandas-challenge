```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

# Player Count

```python
unique_players = purchase_data['SN'].unique()

player_count = pd.DataFrame({'Total Players': [len(unique_players)]})
player_count
```

# Purchasing Analysis (Total)

```python
unique_items = len(purchase_data['Item ID'].unique())
average_profits = round(purchase_data['Price'].mean(),2)
number_purchases = purchase_data.shape[0]
total_revenue = purchase_data['Price'].sum()
pd.options.display.float_format = '${:,.2f}'.format

purchasing_analysis_total = pd.DataFrame({'Number of Unique Items': [unique_items], 
                                          'Average Price': [average_profits],
                                          'Number of Purchases': [number_purchases], 
                                          'Total Revenue': [total_revenue]})
purchasing_analysis_total
```

# Gender Demographics

```python
unique_players2 = purchase_data.drop_duplicates(['SN'])
pd.options.display.float_format = '{:,.2f}%'.format

# checking males stats 
unique_males_only = unique_players2.loc[unique_players2["Gender"] == "Male", :]
unique_males_percentage = round(len(unique_males_only) / len(unique_players2) * 100, 2)

# checking females stats 
unique_females_only = unique_players2.loc[unique_players2["Gender"] == "Female", :]
unique_females_percentage = round(len(unique_females_only) / len(unique_players2) * 100, 2)

# checking unidentified gender stats  
unique_nondisclosed_only = unique_players2.loc[unique_players2["Gender"] == "Other / Non-Disclosed", :]
unique_nondisclosed_percentage = round(len(unique_nondisclosed_only) / len(unique_players2) * 100, 2)

demographics = {'Total Count': [len(unique_males_only), len(unique_females_only), len(unique_nondisclosed_only)], 
                'Percentage of Players': [unique_males_percentage, unique_females_percentage, unique_nondisclosed_percentage]}

gender_demographics = pd.DataFrame(demographics, index = ["Male", "Female", "Other / Non-Disclosed"])
gender_demographics
```

# Purchasing Analysis (Gender)

```python
all_males_only = purchase_data.loc[purchase_data["Gender"] == "Male", :]
total_all_males_purchase = all_males_only['Price'].sum()
avg_all_males_purchase = round(all_males_only['Price'].mean(),2)
avg_unique_males_purchase = round(total_all_males_purchase / len(unique_males_only),2)

all_females_only = purchase_data.loc[purchase_data["Gender"] == "Female", :]
total_all_females_purchase = all_females_only['Price'].sum()
avg_all_females_purchase = round(all_females_only['Price'].mean(),2)
avg_unique_females_purchase = round(total_all_females_purchase / len(unique_females_only),2)

all_others_only = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", :]
total_all_others_purchase = all_others_only['Price'].sum()
avg_all_others_purchase = round(all_others_only['Price'].mean(),2)
avg_unique_others_purchase = round(total_all_others_purchase / len(unique_nondisclosed_only),2)

pd.options.display.float_format = '${:,.2f}'.format
purchasing_analysis = {'Purchase Count': [len(all_females_only), len(all_males_only), len(all_others_only)], 
                       'Average Purchase Price': [avg_all_females_purchase, avg_all_males_purchase, avg_all_others_purchase],
                       'Total Purchase Value': [total_all_females_purchase, total_all_males_purchase, total_all_others_purchase], 
                       'Avg Total Purchase per Person': [avg_unique_females_purchase, avg_unique_males_purchase, avg_unique_others_purchase]}

purchasing_analysis_gender = pd.DataFrame(purchasing_analysis, index = ["Female", "Male", "Other / Non-disclosed"])
purchasing_analysis_gender 
```

# Age Demographics

```python
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 60]
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

age_demographics = pd.DataFrame(columns=['Total Count', 'Percentage of Players'], index=age_labels)
pd.options.display.float_format = '{:,.2f}%'.format

age_demographics['Total Count'] = pd.cut(unique_players2["Age"], age_bins, labels=age_labels).value_counts()
age_demographics['Percentage of Players'] = age_demographics["Total Count"]/len(unique_players2) * 100
age_demographics
```

# Purchasing Analysis (Age)

```python
purchasing_analysis_age = pd.DataFrame(columns=['Purchase Count', 'Average Purchase Price', 'Total Purchase Value', 'Avg Total Purchase per Person'], index=age_labels)
pd.options.display.float_format = '${:,.2f}'.format

purchasing_analysis_age['Purchase Count'] = pd.cut(purchase_data["Age"], age_bins, labels=age_labels).value_counts()
purchase_data['Age ranges'] = pd.cut(purchase_data["Age"], age_bins, labels=age_labels)

grouping_age_ranges = purchase_data.groupby('Age ranges').sum()
purchasing_analysis_age['Total Purchase Value'] = grouping_age_ranges['Price']
purchasing_analysis_age['Average Purchase Price'] = grouping_age_ranges['Price'] / purchasing_analysis_age['Purchase Count'] 
purchasing_analysis_age['Avg Total Purchase per Person'] = grouping_age_ranges['Price'] / age_demographics['Total Count'] 

purchasing_analysis_age
```

# Top Spenders

```python
top_spenders = pd.DataFrame(columns=['Purchase Count', 'Average Purchase Price', 'Total Purchase Value'])
pd.options.display.float_format = '${:,.2f}'.format

top_spenders['Purchase Count'] = purchase_data.groupby(['SN'])['SN'].count()
top_spenders['Total Purchase Value'] = purchase_data.groupby(['SN'])['Price'].sum()
top_spenders['Average Purchase Price'] = round(purchase_data.groupby(['SN'])['Price'].mean(),2)

top_spenders.sort_values("Total Purchase Value", ascending=False).head(5)
```

# Most Popular Items

```python
most_popular_items = pd.DataFrame(columns=['Purchase Count', 'Item Price', 'Total Purchase Value'])
pd.options.display.float_format = '${:,.2f}'.format

most_popular_items['Purchase Count'] = purchase_data.groupby(['Item ID','Item Name'])['Item Name'].count()
most_popular_items['Item Price'] = purchase_data.groupby(['Item ID','Item Name'])['Price'].first()
most_popular_items['Total Purchase Value'] = purchase_data.groupby(['Item ID','Item Name'])['Price'].sum()

most_popular_items.sort_values("Purchase Count", ascending=False).head(5)
```

# Most Profitable Items

```python
most_popular_items = pd.DataFrame(columns=['Purchase Count', 'Item Price', 'Total Purchase Value'])
pd.options.display.float_format = '${:,.2f}'.format

most_popular_items['Purchase Count'] = purchase_data.groupby(['Item ID','Item Name'])['Item Name'].count()
most_popular_items['Item Price'] = purchase_data.groupby(['Item ID','Item Name'])['Price'].first()
most_popular_items['Total Purchase Value'] = purchase_data.groupby(['Item ID','Item Name'])['Price'].sum()

most_popular_items.sort_values("Total Purchase Value", ascending=False).head(5)
```
